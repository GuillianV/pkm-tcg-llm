```python
import requests
import json
from PIL import Image
import os
from rapidfuzz import fuzz
import imagehash
import cv2
import numpy as np
from dotenv import load_dotenv 
from io import BytesIO


load_dotenv()
```

```python

TCG_HASH_FOLDER="data/hash"
TCG_JSON_FOLDER="data/tcg"
TCG_POKEMON_API_KEY = os.getenv("TCG_POKEMON_API_KEY")
```

```python
jsonHash = {}
jsonHashName ="hash_img.json"
hashTotalIndex = 0
```

```python

def tcg_hash_save (data,filename):
    with open(TCG_HASH_FOLDER+"/"+filename,'w') as f:
        json.dump(data,f)

```

```python

hashIndex = 0
for file in os.listdir(TCG_JSON_FOLDER): 
   
    try:
        hashIndex+=1
    
        if hashTotalIndex > hashIndex:
            continue
    
        file_full_path = TCG_JSON_FOLDER +"/"+ file
        data = json.load(open(file_full_path))
        imgpath = data["images"]["large"]
        response = requests.get(imgpath,headers={"X-Api-Key": TCG_POKEMON_API_KEY, 'Accept': 'image/jpg'})
        image = Image.open(BytesIO(response.content))
        image =  image.resize((500,500))
        hash = imagehash.phash(image)
        jsonHash[str(hash)] = data["id"]
        tcg_hash_save(jsonHash,jsonHashName)
        hashTotalIndex = hashIndex
        print(str(hashIndex)+ " : " + data["id"])
        data= None
        response = None
        image = None
        
            
    except Exception as e:
        print(e)
        continue
```

```python

import cv2
import numpy as np
from PIL import Image
import imagehash
import os

def compare_images(image_path, database_path):
    # Charger l'image de la carte découpée
    query_image = Image.open(image_path)
    query_hash = imagehash.phash(query_image)

    # Parcourir la base de données d'images
    closest_match = None
    closest_distance = float('inf')

    for filename in os.listdir(database_path):
        if filename.endswith(('.jpg', '.jpeg', '.png')):
            # Charger l'image de la base de données
            db_image = Image.open(os.path.join(database_path, filename))
            db_hash = imagehash.phash(db_image)

            # Calculer la distance entre les hachages
            distance = query_hash - db_hash

            # Mettre à jour la correspondance la plus proche
            if distance < closest_distance:
                closest_distance = distance
                closest_match = filename

    return closest_match

# Exemple d'utilisation
image_path = '../TCG-Card-Recognizer/test/mew2-d.jpg'
database_path = 'path_to_your_database'
closest_match = compare_images(image_path, database_path)
print(f'La carte la plus proche est : {closest_match}')
# Exemple d'utilisation
```
